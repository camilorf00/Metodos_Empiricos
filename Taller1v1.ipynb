```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from scipy import stats
```

# Fondos de Renta Variable

```python
data = pd.read_excel("DatosTaller1.xlsx", index_col = 0)
data.head()
```

### Modelo Din√°mico CrediCorp

```python
CrediD = data[['Dinamico','COLCAP']].copy()
CrediD = np.log(CrediD) - np.log(CrediD.shift(1))
CrediD['Dinamico'] = CrediD['Dinamico'] - data['RiskFree']
CrediD['COLCAP'] = CrediD['COLCAP'] - data['RiskFree'] + data['DividendYield']
CrediD = CrediD.dropna()
```

```python
Y1 = CrediD['Dinamico']
X1 = CrediD['COLCAP']
X1 = sm.add_constant(X1)
```

```python
modelValor = sm.OLS(Y1,X1).fit()
modelValor.summary()
```
